```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv", skiprows=4)
df = df.iloc[:,:-1]
```

```python
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['2023'], bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel("Population in 2023")
plt.ylabel("Frequency")
plt.title("Distribution of Population in 2023")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df['log_population'] = np.log1p(df['2023'])

# Plot the histogram
plt.figure(figsize=(10,6))
plt.hist(df['log_population'], bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel("Log(Population in 2023)")
plt.ylabel("Frequency")
plt.title("Distribution of Population in 2023 (Log Scale)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
